```python
onstatements = " "
mylist = df.select('names').collect()

for pokemon in mylist:
    if pokemon.names is not None: 
        print(pokemon.names)
        onstatements = onstatements + ' pknames.names = pkattack.'+pokemon.names+ " and"
```
